```python
import pandas as pd

coffee = pd.read_csv('warmup-data/coffee.csv')

coffee['price'] = coffee['Coffee Type'].map({'Espresso': 3.99, 'Latte': 4.99, 'Cappuccino': 5.99})

coffee
```

```python
coffee['revenue'] = coffee['price'] * coffee['Units Sold']
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
coffee['percent_change'] = (coffee['revenue'] - coffee['yesterday_revenue']) / coffee['yesterday_revenue'] * 100

coffee
```

```python
coffee['revenue_rank'] = coffee['revenue'].rank(method='average', ascending=True)

coffee = coffee.sort_values(by='revenue_rank', ascending=False)

coffee
```

```python
coffee.sort_index(inplace=True)

coffee
```

```python
coffee['cumsum'] = coffee['revenue'].cumsum()

coffee
```

```python
coffee['rolling'] = coffee['revenue'].rolling(3).sum()

coffee
```
