```python
import pandas as pd

coffee = pd.read_csv('warmup-data/coffee.csv')

coffee['price'] = coffee['Coffee Type'].map({'Espresso': 3.99, 'Latte': 4.99, 'Cappuccino': 5.99})

coffee['revenue'] = coffee['price'] * coffee['Units Sold']

coffee
```

```python
pivot = coffee.pivot(columns='Coffee Type', index='Day', values='revenue')

pivot
```

```python
pivot.sum(axis=1)
```

```python
pivot.sum(axis=0)
```
