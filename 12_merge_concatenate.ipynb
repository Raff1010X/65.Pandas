```python
import pandas as pd

bios = pd.read_csv('data/bios.csv')

bios.head()
```

```python
nocs = pd.read_csv('data/noc_regions.csv')

nocs.head()
```

https://cdn.educba.com/academy/wp-content/uploads/2020/04/Python-Pandas-Join.jpg.webp

```python
# pd.merge(bios, nocs, on='NOC').head()

bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left', suffixes=['_bios', '_nocs'])

bios_new.head()
```

```python
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)

bios_new.drop(columns=['notes'], inplace=True)

bios_new.head()
```

```python
bios_new[bios_new['NOC_bios'] != bios_new['born_country_full']][['born_date', 'NOC_bios', 'born_country_full']]
```

```python
usa = bios_new[bios_new['born_country'] == 'USA']

fra = bios_new[bios_new['born_country'] == 'FRA']

usa_fra = pd.concat([usa, fra])

usa_fra
```

```python
results = pd.read_csv('data/results.csv')

results.head()
```

```python
merged = pd.merge(results, bios_new, on='athlete_id', how='left').filter(['year', 'type', 'discipline', 'event', 'as', 'name', 'medal'])

merged = merged.sort_values(by=['medal', 'year'], ascending=[False, False])

merged.head(10)
```
