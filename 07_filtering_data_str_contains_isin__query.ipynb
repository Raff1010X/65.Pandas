```python
import pandas as pd

pd.__version__
```

```python
bios = pd.read_csv('data/bios.csv')

bios.tail(3)
```

```python
bios.info()
```

```python
results = bios.loc[bios['height_cm'] >= 220, ['height_cm', 'name']]
results = result.sort_values(by='height_cm', ascending=False)
results = result.head(10)

results
```

```python
bios[bios['height_cm'] >= 220][['height_cm', 'name']].sort_values(by='height_cm', ascending=False).head(10)
```

```python
bios[(bios['height_cm'] >= 215) & (bios['born_country'] == 'USA')].filter(items=['height_cm', 'name']).sort_values(by='height_cm', ascending=False).head(10)
```

```python
results = bios.loc[(bios['height_cm'] >= 215) & (bios['born_country'] == 'USA')]
results = results.filter(items=['height_cm', 'name'])
results = results.sort_values(by='height_cm', ascending=False)
results = results.head(10)

results
```

```python
bios[bios['name'].str.contains('O\'neal|robinson', na=False, case=False)].sort_values(by='born_date', ascending=True).head(10).filter(items=['athlete_id', 'name', 'born_date', 'born_city', 'born_country'])
```

```python
results = bios[bios['name'].isin(["Ross Robinson", "Sid Robinson"]) & (bios['born_country'] != 'USA')]
results = results.sort_values(by='born_date', ascending=True)
results = results.head(10)
results = results.filter(items=['athlete_id', 'name', 'born_date', 'born_city', 'born_country'])

results
```

```python
bios.query('born_country == "USA" and height_cm >= 215').sort_values(by='born_date', ascending=True).head(10).filter(items=['athlete_id', 'name', 'born_date', 'born_city', 'born_country'])
```
