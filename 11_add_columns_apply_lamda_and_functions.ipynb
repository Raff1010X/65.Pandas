```python
import pandas as pd

bios = pd.read_csv('data/bios.csv')

bios.head()
```

```python
bios.groupby('height_cm')['height_cm'].count().sort_values(ascending=False)
```

```python
# add new column "height category" using lambda function

# height category lamda function

height_category = lambda x: 'undefined' if pd.isnull(x) else ('short' if x < 170 else ('average' if x < 190  else 'tall') )

bios['height_category'] = bios['height_cm'].apply(height_category)

bios[['height_cm', 'height_category']].head(10)
```

```python
bios.groupby('weight_kg')['weight_kg'].count().sort_values(ascending=False)
```

```python
def mass_category(row):
    if pd.isnull(row['weight_kg']):
        return 'undefined'
    elif row['weight_kg'] < 80:
        return 'light'
    elif row['weight_kg'] < 100:
        return 'average'
    elif row['weight_kg'] < 120:
        return 'heavy'
    else:
        return 'very heavy'

bios['mass_category'] = bios.apply(mass_category, axis=1)

bios[['weight_kg', 'mass_category']].sample(10)
```
