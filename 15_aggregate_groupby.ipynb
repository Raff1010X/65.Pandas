```python
import pandas as pd

coffee = pd.read_csv('warmup-data/coffee.csv')

coffee['price'] = coffee['Coffee Type'].map({'Espresso': 3.99, 'Latte': 4.99, 'Cappuccino': 5.99})

coffee
```

```python
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'price': 'sum'})
```

```python
coffee.assign(Total_Revenue=lambda x: x['Units Sold'] * x['price']).groupby(['Day']).agg({'Units Sold': 'sum', 'Total_Revenue': 'sum'})
```

```python
coffee.groupby(['Coffee Type', 'Day']).agg({'Units Sold': 'sum', 'price': 'sum'})
```

```python
coffee.assign(Total_Revenue=lambda x: x['Units Sold'] * x['price']).groupby(['Coffee Type', 'Day']).agg({'Units Sold': 'sum', 'Total_Revenue': 'sum'})
```
