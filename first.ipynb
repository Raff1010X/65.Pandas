```python
import pandas as pd
import sqlite3

# create a sample DataFrame
df = pd.DataFrame({
    'Name': ['John', 'Mary', 'David', 'Emily'],
    'Age': [25, 31, 42, 28]
})

# create a SQLite database in memory
conn = sqlite3.connect(':memory:')

# write the DataFrame to the database
df.to_sql('people', conn, index=False)

# execute a SQL query on the DataFrame
cur = conn.cursor()
cur.execute('SELECT * FROM people WHERE Age > 30 AND Name NOT LIKE "David"')

# fetch the results
results = cur.fetchall()

# create a new DataFrame with the results
results_df = pd.DataFrame(results, columns=df.columns)

# print the results DataFrame
print(results_df)

# close the connection
conn.close()
```

```python
import pandas as pd
from pandasql import sqldf

# create a sample DataFrame
df = pd.DataFrame({
    'Name': ['John', 'Mary', 'David', 'Emily', 'Anna'],
    'Age': [25, 31, 12, 28, 11],
    'Sex': ['M', 'F', 'M', 'F', 'F']
})

# execute a SQL query on the DataFrame
results = sqldf('''
                SELECT 
                    Płec,
                    AVG(Age) AS "Średni wiek"
                FROM (
                    SELECT 
                        CASE 
                            WHEN Sex = 'M' AND Age > 18 THEN 'Meżczyźni'
                            WHEN Sex = 'F' AND Age > 18 THEN 'Kobiety'
                            ELSE 'Dzieci'
                        END AS "Płec",
                        Age
                    FROM df
                ) AS subquery
                GROUP BY Płec
                 ''')

# print the results
print(results)

import numpy as np

# create a new column 'Płec' based on the conditions
df['Płec'] = np.where((df['Sex'] == 'M') & (df['Age'] > 18), 'Meżczyźni',
                     np.where((df['Sex'] == 'F') & (df['Age'] > 18), 'Kobiety', 'Dzieci'))

# group by 'Płec' and calculate the average age
results2 = df.groupby('Płec')['Age'].mean().reset_index()

# rename the columns
results2.columns = ['Płec', 'Średni wiek']

# print results2
print(results2)

```
